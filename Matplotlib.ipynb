```python
 import pandas as pd
```

```python
s=pd.read_csv("https://people.sc.fsu.edu/~jburkardt/data/csv/ford_escort.csv")
```

```python
print(s[s.columns])
```

```python
#ploting the histogram
h=(s[s.columns[2]])
m=(s[s.columns[1]])
plt.hist(h,histtype='bar',rwidth=0.8,label="Price")
plt.hist(m,histtype='bar',rwidth=0.8,label="mileage")
plt.xlabel('x')
plt.ylabel('y')
plt.title("histogram")
plt.legend()
```

```python
#plt the scatter plot

#ploting the histogram
h=(s[s.columns[2]])
m=(s[s.columns[0]])
plt.scatter(h,m,label="Price",color="green")
plt.xlabel('price')
plt.ylabel('year')
plt.title("histogram")
plt.legend()
```

```python
#ploting the stack 
for i in s.columns:
        print(i)
plt.stackplot(s[s.columns[0]],s[s.columns[1]])
```

```python
import numpy as np
days=np.arange(1,5)
sleep=np.arange(1,5)
walking=np.arange(1,5)
playing=np.arange(1,5)
#to make the legends we will add empty lines 
plt.plot([],[],color='m',label='sleeping')
plt.plot([],[],color='k',label='walking')
plt.plot([],[],color='c',label='playing')

plt.stackplot(days,sleep,walking,playing,colors=['m','k','c'])
plt.legend()
```
